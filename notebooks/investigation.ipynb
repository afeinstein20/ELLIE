```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import os, sys
from ellie import visualize
from bokeh.io import show
```

```python
fns = np.array(os.listdir())
inds = np.array([i for i,item in enumerate(fns) if '.fits' in item])
fns = fns[inds]
```

```python
fig, axes = plt.subplots(nrows=3, ncols=6, figsize=(20,10))
i = 0
for ax in axes.reshape(-1):
    hdu = fits.open(fns[i])
    tpf = hdu[0].data
    title = str(hdu[0].header['TMAG']) + ' ' + str(hdu[0].header['postcard'])
    ax.imshow(tpf, origin='lower', cmap='viridis', vmin=60, vmax=300)
    ax.set_title(title)
    i += 1
plt.show()
```
